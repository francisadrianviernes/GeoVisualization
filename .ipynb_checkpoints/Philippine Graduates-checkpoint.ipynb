```python
provinces["Region"] = ""
#Store the provinces column for empty dataframe later to append data to
cols = provinces.columns
```

```python
#Regional Provinces List
CALABARZON = ["Batangas", "Cavite", "Quezon", "Rizal"]
Central_Luzon = ['Aurora', "Bataan", "Bulacan", "Nueva Ecija", "Pampanga", 'Tarlac', "Zambales"]
Central_Visayas = ["Bohol", "Cebu", "Negros Oriental", "Siquijor", "Negros Oriental"]
Western_Visayas = ["Aklan", "Antique","Capiz", "Guimaras", "Iloilo", "Negros Occidental"]
Bicol_Region = [ "Albay", "Camarines Norte", "Camarines Sur","Sorsogon", "Catanduanes", "Masbate"]
Ilocos_Region = ["Ilocos Norte","Ilocos Sur", "La Union", "Pangasinan"]
Northern_Mindanao = ["Bukidnon", "Camiguin", "Misamis Occidental", "Misamis Oriental", "Lanao del Norte"]
Davao_Region = ['Compostela Valley', 'Davao del Norte', 'Davao del Sur', 'Davao Oriental']
Eastern_Visayas = [ 'Biliran', 'Leyte', 'Northern Samar', 'Samar', 'Eastern Samar', 'Southern Leyte', 'Ormoc']
SOCCSKSARGEN = ['South Cotabato', 'North Cotabato', 'Sultan Kudarat', 'Sarangani', 'General Santos']
Zamboanga_Peninsula = ['Zamboanga del Norte','Zamboanga Sibugay', 'Zamboanga del Sur']
Cagayan_Valley = ['Batanes', "Cagayan", "Isabela", "Nueva Vizcaya", "Quirino"]
MIMAROPA = ["Occidental Mindoro", "Oriental Mindoro", "Marinduque", "Romblon", "Palawan"]
Cordillera_Adm_Region = ['Abra', 'Apayao', 'Benguet', 'Ifugao', 'Kalinga', 'Mountain Province']
CARAGA = [ 'Agusan del Norte', 'Agusan del Sur', 'Dinagat Islands', 'Surigao del Norte', 'Surigao del Sur']
ARMM = [ 'Basilan', 'Lanao del Sur', 'Maguindanao', 'Sulu', 'Tawi-Tawi']

# provinces[provinces.NAME_1.isin(Davao_Region)]
# provinces[provinces.NAME_1.str.contains("Mindoro")]
```

```python
final_df = pd.DataFrame(columns=cols)


provinces_list = [
    ["Metropolitan Manila"],
    CALABARZON,
    Central_Luzon,
    Central_Visayas,
    Western_Visayas,
    Bicol_Region,
    Ilocos_Region,
    Northern_Mindanao,
    Davao_Region,
    Eastern_Visayas,
    SOCCSKSARGEN,
    Zamboanga_Peninsula,
    Cagayan_Valley,
    MIMAROPA,
    Cordillera_Adm_Region,
    CARAGA,
    ARMM   
]

region_list = [
    "National Capital Region",
    "CALABARZON",
    "Central Luzon",
    "Central Visayas",
    "Western Visayas",
    "Bicol Region",
    "Ilocos Region",
    "Northern Mindanao",
    "Davao Region",
    "Eastern Visayas",
    "SOCCSKSARGEN",
    'Zamboanga Peninsula',
    'Cagayan Valley',
    'MIMAROPA',
    'Cordillera Adm. Region',
    'CARAGA',
    'ARMM'
]

print("Length of provinces list: ", len(provinces_list))
print("Length of region list: ", len(region_list))


for i in range(len(provinces_list)):
    for j in provinces_list[i]:
        temp_df = provinces[provinces.NAME_1==j]
        temp_df["Region"] = region_list[i]
        final_df = pd.concat([final_df,temp_df])
```

```python
merged = final_df.merge(df, on="Region")
merged = gpd.GeoDataFrame(merged)
```

```python
## Merging Raw Data
df2 = pd.read_excel(r'D:\Downloads\04.15.20_Enrolles and Graduates_final version.xlsx',
                   sheet_name='Sheet1')
```

```python
#Filter school year and status
df2 = df2[(df2.school_year=="2017-18")&(df2.status=="Graduates")]
```

```python
df2
```

```python
# df2.groupby(["Region", 'PSCED Name']).sum()
df3 = df2[["Region", "PSCED Name", "Value"]]
df3 = df3.groupby(["Region", "PSCED Name"]).sum().sort_values(ascending=False).reset_index()
df3 = df3.drop_duplicates(subset=["Region"], keep="first").reset_index(drop=True)
df3 = df3.rename(columns={0: "Graduates"})
df3 = df3[["Region", "PSCED Name", "Graduates"]]
```

```python
df3
```

```python
df3.iat[0,0] = "National Capital Region"
df3.iat[11,0] = "SOCCSKSARGEN"
df3.iat[14,0] = "CARAGA"
```

```python
top_program = gpd.GeoDataFrame(df3.merge(merged, on="Region"))
```

```python
fig, ax = plt.subplots(1,1, figsize=(20,20))
top_program.plot(column='PSCED Name', categorical=True, cmap='Spectral', linewidth=.6, edgecolor='0.2',
         legend=True, legend_kwds={'bbox_to_anchor':(.3, 1.05),'fontsize':16,'frameon':False}, ax=ax)
ax.axis('off')
ax.set_title('Top Course of Graduates per Region',fontsize=20);

```
