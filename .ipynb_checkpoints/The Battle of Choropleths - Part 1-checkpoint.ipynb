# PRELIMINARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors


import geopandas as gpd

%matplotlib inline
```

# LOAD AND PREPROCESS THE DATA

```python
df = pd.read_csv('data/gdp_per_capita.csv', 
            skiprows=4)
df = df.loc[:,['Country Name','Country Code', '2020']] #Choose only 2020
df.head()
```

## LOAD SHAPEFILES

```python
gdf = gpd.read_file('shapefiles/world-administrative-boundaries/world-administrative-boundaries.shp')
gdf.head()
```

```python
gdf.loc[gdf.name=="Philippines", 'geometry'][202]
```

## MERGE DATA

```python
merged = gdf.merge(df, left_on='name', right_on='Country Name' )
merged.head()
```

As we can see, not all names are the same in our two dataframes. We need to make some more efforts to uncovering these and after further cleaning:

```python
merged.fillna(0, inplace=True)
```

# PLOTTING

```python
cmap = 'Greens'

#Create A Figure to Attach Other Objects Later
fig, ax = plt.subplots(figsize = (30,25))

merged.plot(column="2020",
            ax=ax,
                 cmap=cmap, 
#                      k=colors,
                 scheme='quantiles',
                 legend=True)

ax.axis('off')

```

## CREATE THE COLORBAR

```python
# create the colorbar
norm = colors.Normalize(vmin=merged['2020'].min(), vmax=merged['2020'].max())
cbar = plt.cm.ScalarMappable(norm=norm, cmap='Greens')
```

```python
cax = fig.add_axes([1, 0.1, 0.03, 0.8])
cbr = fig.colorbar(cbar, cax=cax,)

#Remove the legend
ax.get_legend().remove()


#Change the tick for the colorbar
cax.tick_params(labelsize=18)

fig
```

```python
#Try a diverging colormap
cmap = 'RdYlGn'

#Create A Figure to Attach Other Objects Later
fig, ax = plt.subplots(figsize = (30,25))

merged.plot(column="2020",
            ax=ax,
                 cmap=cmap, 
#                      k=colors,
                 scheme='quantiles',
                 legend=True)

# ax.legend(loc="best")

ax.axis('off')
```

```python
# create the colorbar
norm = colors.Normalize(vmin=merged['2020'].min(), vmax=merged['2020'].max())
cbar = plt.cm.ScalarMappable(norm=norm, cmap=cmap)

cax = fig.add_axes([1, 0.1, 0.03, 0.8])
cbr = fig.colorbar(cbar, cax=cax,)

#Remove the legend
ax.get_legend().remove()


#Change the tick for the colorbar
cax.tick_params(labelsize=18)

fig
```

## PUT A LABEL ON THOSE WITH ABOVE 100,000 USD GDP per Capita

```python
for ind, row in merged.iterrows():
    if merged.loc[ind,'2020']>100000:
        ax.text(row["geometry"].centroid.x,row["geometry"].centroid.y+4, row["name"], fontsize=12, color='blue',weight="bold")
        ax.text(row["geometry"].centroid.x,row["geometry"].centroid.y, row["2020"],
                fontsize=12, color='blue', weight='bold')

fig
# fig = plt.gcf()    

# # fig.savefig('GDP PER CAPITA OF THE WORLD.png', dpi=1200)
```
